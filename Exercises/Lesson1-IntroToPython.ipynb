###<font color='red'> NumPy Arrays vs Python Lists


````python
%%time
import time
import math as m

#Step-1: define the number of elements in the list
nelements = 1000000

#Step-2: Create the list and store integer the elements in it
a = [i for i in range(nelements)]

#Step-3: Calculate something involving the elements of a
for element in a:
  x = element/(nelements-1) * 2 * m.pi
  y = m.sin(x)
````

If we want to save the elements, we need to walk the list element by element and define the list using list comprehension:

````python
x = [i/(nelements-1) * 2 * m.pi for i in a]
y = [m.sin(i) for i in x]
````

````python
x = []
y = []
for element in a:
  xtemp = element/(nelements-1) * 2 * m.pi
  x.append(xtemp)
  y.append(m.sin(xtemp))
````

Let's see now what happens with Numpy

````python
%%time
import numpy as np

a = np.arange(0, nelements)
x = a/(nelements-1)*2*np.pi
y = np.sin(x)
````

We see two things:

1.   Compared with native Python, NumPy is much faster.
2.   Array operations are vectorized. The desired operation is applied to all elements at once. Therefore, there is not need to append elements to a list

###<font color='red'> Array Indexing and Slicing


Unlike lists, Numpy arrays support Indexing from lists and arrays. For example

````python
#Define an arbitrary list
mylist = [1,3,4,5,6]

#Define the index we would like to retrieve
idx = [0,1,3]

#Try to access the list from the indices. Produces an error
print(mylist[idx])
````

````python
#Define an arbitrary Numpy Array
myarray = np.array([1,3,4,5,6])

#Define an index list
idx = [0,1,3]

#Access the elements of myarray using the desired indices
print(myarray[idx])

#This can also be done if idx is an array
idx = np.array([0,1,3])
print(myarray[idx])
````

Like lists, Numpy arrays can also be sliced. For example:

````python
print("List slicing")

mylist = [1,3,4,5,6]
print(mylist[:2])
print(mylist[2:])
print(mylist[2:4])
print(mylist[::2])
print(mylist[::-1])

print("Array slicing")
myarray = np.array([1,3,4,5,6])
print(myarray[:2])
print(myarray[2:])
print(myarray[2:4])
print(myarray[::2])
print(myarray[::-1])
````

###<font color='red'> Array Math

Unlike Python lists, Numpy arrays support a number of mathematical operations, which are generally applied element by element.

````python
import numpy as np

a = np.array([0,2,4,6,8])
b = np.array([1,3,5,7,9])

print("a = ", a)
print("b = ", b)
print("a+b = ", a+b)
print("a-b = ", a-b)
print("a*b = ", a*b)
print("a/b = ", a/b)
print("a**b = ", a**b)
````

These operations are equivalent to using the NumPy methods

````python
import numpy as np

a = np.array([0,2,4,6,8])
b = np.array([1,3,5,7,9])

print("a = ", a)
print("b = ", b)
print("a+b = ", np.add(a,b))
print("a-b = ", np.subtract(a,b))
print("a*b = ", np.multiply(a,b))
print("a/b = ", np.divide(a,b))
print("a**b = ", np.power(a,b))
````

The most basic way of producing a plot using matplotlib is the following

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.array([1,2,3,4])
y = np.array([1,4,2,3])

# Create a figure containing a single Axes.
fig, ax = plt.subplots()

# Plot some data on the Axes.
ax.plot(x,y)

# Show the figure.
plt.show()
````

There are several ways of improving this plot. For example, by showing the available data points with a symbol, making the interpolating line "dashed", adding labels to the axes, adding a legend, adding minor ticks, etc...

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.array([1,2,3,4])
y = np.array([1,4,2,3])

# Create a figure containing a single Axes.
fig, ax = plt.subplots()

# Plot some data on the Axes, making sure the data is plotted with large circles (marker='o'; markersize=10), joint together by a dashed line (linestyle='dashed')
# with a relatively thick line (linewidth=2), red (color='red'). The dataset is labelled Data points (label='Data Points', which is used later on for the legend)
ax.plot(x,y, marker='o', linestyle='--', linewidth=2, color='red', markersize=10, label='Data Points-Linear Interpolation')

# Set the x and y lables, with fontsize=14, in inches.
ax.set_xlabel('X [km]', size=14)
ax.set_ylabel('X [km]', size=14)

# Set minor ticks for the plot
ax.minorticks_on()

# Make the legend visible and put at the bottom right part of the plot
ax.legend(loc='lower right')

# Make the tick labels larger.
ax.tick_params(axis='both', which='major', labelsize=12)

# Show the figure.
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.array([1,2,3,4])
y = np.array([1,4,2,3])

# Create a figure containing a single Axes.
fig, ax = plt.subplots()


# Plot some data on the Axes, making sure the data is plotted with large circles (marker='o'; markersize=10), joint together by a dashed line (linestyle='dashed', drawstyle='steps-mid')
# with a relatively thick line (linewidth=2), red (color='red'). The dataset is labelled Data points (label='Data Points', which is used later on for the legend)
ax.plot(x,y, marker='o', linestyle='--', linewidth=2, color='blue', markersize=10, label='Data Points-Histogram', drawstyle='steps-mid')

# Set the x and y lables, with fontsize=14, in inches.
ax.set_xlabel('X [km]', size=14)
ax.set_ylabel('X [km]', size=14)

# Set minor ticks for the plot
ax.minorticks_on()

# Make the legend visible and put at the bottom right part of the plot
ax.legend(loc='lower right')

# Make the tick labels larger.
ax.tick_params(axis='both', which='major', labelsize=12)

# Show the figure.
plt.show()
````

Figures and axes can also be defined manually:

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.array([1,2,3,4])
y = np.array([1,4,2,3])

fig = plt.figure(1, figsize=(6, 3))
ax = fig.add_subplot(111)

ax.plot(x,y, marker='*', linestyle='-.', linewidth=1, color='C00', markersize=10, label='Data Points-Histogram', drawstyle='steps-mid')

#We can now access methods that affect the way axes look.
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Colors are usually defined using an hexadecimal base: You can check the hexadecimal code online (e.g., https://htmlcolorcodes.com/)
ax.set_facecolor('#a8ffff')

plt.show()
````

### <font color='red'> Plotting datasets spanning several orders of magnitude </font>

In Astronomy, it is common that measurments span several orders of magnitude, and so, linearly-spaced plots are not useful.

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.array([1,1e1,1e2,1e3,1e4,1e5,1e6,1e7,1e8,1e9])
y = np.random.rand(len(x))

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
ax[0].plot(x,y, 'o', ls='-')

ax[1].plot(x,y, 'o', ls='-')
ax[1].set_xscale('log')

ax[2].plot(np.log10(x),y, 'o', ls='-')

ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')

ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')

ax[2].set_xlabel(r'$\log_{10}(X)$')
ax[2].set_ylabel('Y')
````

### <font color='red'> Basic numerical integration using SciPy </font>

````python
from scipy.integrate import trapezoid

def quadratic_function(x):
  return x**2

x = np.linspace(0,1,10)
y = quadratic_function(x)

# We can provide the points where the function is evaluated.
integral = trapezoid(y, x)
print("Numerical integral=%.4f, Analytic integral=%.4f, Error=%.3f %%"%(integral, 1./3., (integral-1./3.)/(1./3)*100))

# Alternatively, we can define the interval length between consecutive evaluations.
dx = x[1]-x[0]
integral = trapezoid(y, dx=dx)
print("Numerical integral=%.4f, Analytic integral=%.4f, Error=%.3f %%"%(integral, 1./3., (integral-1./3.)/(1./3)*100))
````

### <font color='red'> Image Manipulation </font>

Scipy includes useful routines for image manipulation. These are included in `scipy.ndimage`. To test these, we can rely on datasets accessible via `scipy.datasets`

````python
from scipy import ndimage
from scipy import datasets

# datasets contain an image we will use. We can also load images from the disk. We will see this later on.
face = datasets.face()
plt.imshow(face)
plt.show()
````

The image is a multidimensional array containing multiple color channels:

````python
print(np.shape(face))
````

We see that the image has 768 pixels in the vertical direction, 1024 pixels in the horizontal direction, and 3 color channels. The red (R), green (G), and blue (B) channels are represented as follows:

* `face[:,:,0]` corresponds to the red (R) channel.
* `face[:,:,1]` corresponds to the green (G) channel.
* `face[:,:,2]` corresponds to the blue (B) channel.

The `plt.imshow` function interprets these RGB channels and displays a colored image when they are present.

The code below demonstrates how to plot each of these channels:

````python
from scipy import ndimage
from scipy import datasets

face = datasets.face()
fig,axes = plt.subplots(ncols=3, nrows=1, figsize=(15,5))
labels = ['R','G', 'B']

for i in range(3):
  axes[i].imshow(face[:,:,i], cmap='Greys', vmin=0, vmax=255)
  axes[i].set_title(labels[i])

plt.show()
````
